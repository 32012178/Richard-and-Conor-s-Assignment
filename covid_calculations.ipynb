```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
# turn data into base data frame

hospitilisations_raw_df = pd.read_csv('hospitilisations_csv.csv')

hospitilisations_raw_df.head()
```

could find important dates where vaccines became administered and compare a spike before that to after for rate of icu vs general patients


```python
# table of ICU occupancy and general hospital occupancy to compare by

icu_occupancy = hospitilisations_raw_df[["date", "type", "value"]].copy()

icu_occupancy.drop(icu_occupancy.loc[icu_occupancy['type']== "Daily ICU occupancy"].index, inplace=True)
icu_occupancy.drop(icu_occupancy.loc[icu_occupancy['type']== "Daily hospital occupancy"].index, inplace=True)
icu_occupancy.drop(icu_occupancy.loc[icu_occupancy['type']== "Daily hospital occupancy per million"].index, inplace=True)
icu_occupancy = icu_occupancy.rename(columns={'value': 'Daily ICU occupancy per million'})
#icu_occupancy =  icu_occupancy["date", "Daily ICU occupancy per million"]
icu_occupancy.plot.line()

hospital_occupancy = hospitilisations_raw_df[["date", "type", "value"]].copy()

hospital_occupancy.drop(hospital_occupancy.loc[hospital_occupancy['type']== "Daily ICU occupancy"].index, inplace=True)
hospital_occupancy.drop(hospital_occupancy.loc[hospital_occupancy['type']== "Daily hospital occupancy"].index, inplace=True)
hospital_occupancy.drop(hospital_occupancy.loc[hospital_occupancy['type']== "Daily ICU occupancy per million"].index, inplace=True)
hospital_occupancy = hospital_occupancy.rename(columns={'value': 'Daily hospital occupancy per million'})
#icu_occupancy =  icu_occupancy["date", "Daily ICU occupancy per million"]

hospital_occupancy.plot.line()
```
