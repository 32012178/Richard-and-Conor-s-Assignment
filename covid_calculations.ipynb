```python
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt

```

```python
# deaths over time for each age group

deaths_age_raw_df = pd.read_csv('Deaths by Age.csv')
deaths_age = deaths_age_raw_df[["Date", "0-9", "Oct-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70-79", "80-89", "90+"]].copy()
deaths_age = deaths_age.rename(columns={'Oct-19': '10-19'})

#clean the data 
deaths_age.tail(20)

deaths_age['90+'] = deaths_age['90+'].astype(str).str.replace(',', '').astype(int)
deaths_age['80-89'] = deaths_age['80-89'].astype(str).str.replace(',', '').astype(int)
deaths_age['70-79'] = deaths_age['70-79'].astype(str).str.replace(',', '').astype(int)
deaths_age['60-69'] = deaths_age['60-69'].astype(str).str.replace(',', '').astype(int)
deaths_age['50-59'] = deaths_age['50-59'].astype(str).str.replace(',', '').astype(int)
deaths_age['40-49'] = deaths_age['40-49'].astype(str).str.replace(',', '').astype(int)
deaths_age['30-39'] = deaths_age['30-39'].astype(str).str.replace(',', '').astype(int)
deaths_age['20-29'] = deaths_age['50-59'].astype(str).str.replace(',', '').astype(int)
deaths_age['10-19'] = deaths_age['40-49'].astype(str).str.replace(',', '').astype(int)
deaths_age['0-9'] = deaths_age['30-39'].astype(str).str.replace(',', '').astype(int)

#plot the data
deaths_age.plot(x='Date', y=["0-9", "10-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70-79", "80-89", "90+"], kind='line')
plt.title("number of deaths by age group")
plt.ylabel('DEATHS')
plt.xticks(rotation = 45)

f = plt.figure()
f.set_figwidth(100)
f.set_figheight(20)

plt.show()
```

```python
#vaccinations by age over time data frame

#sort for the columns needed
vaccinations_age_raw_df = pd.read_csv("Vaccinations By Age Group and State First.csv")
vaccinations_age = vaccinations_age_raw_df[["Date", "Dec-15", "16-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70-79", "80-89", "90+"]].copy()

#clean the data
vaccinations_age = vaccinations_age.replace({'%':''}, regex=True)
vaccinations_age["Dec-15"] = vaccinations_age["Dec-15"].replace({'-':'0'}, regex=True)
vaccinations_age = vaccinations_age.rename(columns={'Dec-15': '11-15'})

vaccinations_age["11-15"] = pd.to_numeric(vaccinations_age["11-15"])
vaccinations_age["16-19"] = pd.to_numeric(vaccinations_age["16-19"])
vaccinations_age["20-29"] = pd.to_numeric(vaccinations_age["20-29"])
vaccinations_age["30-39"] = pd.to_numeric(vaccinations_age["30-39"])
vaccinations_age["40-49"] = pd.to_numeric(vaccinations_age["40-49"])
vaccinations_age["50-59"] = pd.to_numeric(vaccinations_age["50-59"])
vaccinations_age["60-69"] = pd.to_numeric(vaccinations_age["60-69"])
vaccinations_age["70-79"] = pd.to_numeric(vaccinations_age["70-79"])
vaccinations_age["80-89"] = pd.to_numeric(vaccinations_age["80-89"])
vaccinations_age["90+"] = pd.to_numeric(vaccinations_age["90+"])

#display of data for plot
vaccinations_age.head()
```

```python
#plot the vaccination data frame
x_axis = vaccinations_age["Date"]
y_axis_cols = vaccinations_age[["11-15", "16-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70-79", "80-89", "90+"]]

vaccinations_age.plot(x='Date', y=["11-15", "16-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70-79", "80-89", "90+"], kind='line')

plt.xlabel("date")
plt.ylabel("% of people vaccinated")
plt.title("% of vaccinated australians by age")
plt.xticks(rotation = 45)
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(6)
plt.show()

```

```python
# import data for ICU and hospitilisations
hospitilisations_df_raw = pd.read_csv('Australia (1).csv')

#clean the data
hospitilisations_df_raw = hospitilisations_df_raw.fillna(0)
hospitilisations_df_raw = hospitilisations_df_raw.replace({',':''}, regex=True)
hospitilisations_df_raw["Hospitalisations"] = pd.to_numeric(hospitilisations_df_raw["Hospitalisations"])

#plot the data 
hospitilisations_df_raw.plot(x='Date', y=["Hospitalisations", "ICU"], kind='line')
plt.xlabel("date")
plt.ylabel("number of people")
plt.title("people in icu and hospital")
plt.xticks(rotation = 45)
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(6)
plt.show()
```
